## Merging images from different folders

```
import os
import shutil
import uuid
```

```
# Defining the base directory
base_dir = '/content/images'

# Defining the source folders for drone and bird images
drone_folders = ['drone', 'flying drones', 'drones flying in the sky', 'quadcopter drone']
bird_folders = ['fling birds', 'birds in the sky', 'birds flying in the sky', 'birds soaring in the sky']

# Defineing the destination folders for drones and birds
drones_dest = os.path.join(base_dir, 'drones')
birds_dest = os.path.join(base_dir, 'birds')

# Creating destination folders
os.makedirs(drones_dest, exist_ok=True)
os.makedirs(birds_dest, exist_ok=True)

# Moving images from drone related folders to drones destination folder
for folder in drone_folders:
    source_folder = os.path.join(base_dir, folder)
    for file in os.listdir(source_folder):
        # Generate a unique name for each image using UUID
        new_name = str(uuid.uuid4()) + os.path.splitext(file)[1]
        shutil.move(os.path.join(source_folder, file), os.path.join(drones_dest, new_name))

# Move images from bird relatd folders to birds destination folder
for folder in bird_folders:
    source_folder = os.path.join(base_dir, folder)
    for file in os.listdir(source_folder):
        # Generate a unique name for each image using UUID
        new_name = str(uuid.uuid4()) + os.path.splitext(file)[1]
        shutil.move(os.path.join(source_folder, file), os.path.join(birds_dest, new_name))

print("Images have been moved successfully.")
```



*   Now all the images are stored in their respective class folders: drones and birds.



```
# Defining the paths to the drones and birds folders
drones_dir = os.path.join(base_dir, 'drones')
birds_dir = os.path.join(base_dir, 'birds')

# Counting the number of images in the drones folder
num_drones_images = len([name for name in os.listdir(drones_dir) if os.path.isfile(os.path.join(drones_dir, name))])

# Counting the number of images in the birds folder
num_birds_images = len([name for name in os.listdir(birds_dir) if os.path.isfile(os.path.join(birds_dir, name))])

# Printing the images in both class folders
print("Number of images in the drones folder:", num_drones_images)
print("Number of images in the birds folder:", num_birds_images)
```

## Data Partitioning

```
# Defining the base directory
base_dir = '/content/images'

# Defining the names of the partition folders
folders = ['train', 'validation', 'test']

# Creating train, validation, and test folders
for folder in folders:
    folder_path = os.path.join(base_dir, folder)
    os.makedirs(folder_path, exist_ok=True)

    # Creating drones and birds subfolders in each partition
    drones_folder = os.path.join(folder_path, 'drones')
    birds_folder = os.path.join(folder_path, 'birds')
    os.makedirs(drones_folder, exist_ok=True)
    os.makedirs(birds_folder, exist_ok=True)

# Creating a function to copy images from source to destination
def copy_images(source_folder, destination_folder, num_images):
    images = os.listdir(source_folder)[:num_images]
    for image in images:
        source_path = os.path.join(source_folder, image)
        destination_path = os.path.join(destination_folder, image)
        shutil.copyfile(source_path, destination_path)

# Copying images to train folder
copy_images(os.path.join(base_dir, 'drones'), os.path.join(base_dir, 'train', 'drones'), 100)
copy_images(os.path.join(base_dir, 'birds'), os.path.join(base_dir, 'train', 'birds'), 100)

# Copying images to validation folder
copy_images(os.path.join(base_dir, 'drones'), os.path.join(base_dir, 'validation', 'drones'), 50)
copy_images(os.path.join(base_dir, 'birds'), os.path.join(base_dir, 'validation', 'birds'), 50)

# Copying images to test folder
copy_images(os.path.join(base_dir, 'drones'), os.path.join(base_dir, 'test', 'drones'), len(os.listdir(os.path.join(base_dir, 'drones'))) - 150)
copy_images(os.path.join(base_dir, 'birds'), os.path.join(base_dir, 'test', 'birds'), len(os.listdir(os.path.join(base_dir, 'birds'))) - 150)

print("Images have been copied to train, validation, and test folders.")
```

# Modelling: Convolutional Neural Network

```
#importing required libraries for cnn modelling
from tensorflow import keras
from keras import layers
from keras import models
```

## Model 1 (without Dropout & Data Augmentation)

### Building the CNN Model

```
# building the CNN model
model = models.Sequential() #Initializing the model using sequential API
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(180, 180, 3))) #Input shape of 180x180 rgb images
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid')) # output node with sigmoid activation for classification
```

```
model.summary()# printing model summary for outshape and parameters of each layer
```

### Compiling the CNN Model

```
from keras import optimizers

# compiling the model
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])
```

### Data Preprocessing

```
from keras.preprocessing.image import ImageDataGenerator

# Rescaling all images by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Defining the train directory using the base directory
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 180x180
        target_size=(180, 180),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(180, 180),
        batch_size=20,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(180, 180),
        batch_size=20,
        class_mode='binary')
```

#### Checking shape of preprocessed data

```
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```



*    The data batch shape indicates that there are 20 images in the batch, each with dimensions of 180 x 180 pixels and 3 channels (RGB).

* The labels batch shape indicates that there are 20 corresponding labels for the images in the batch.



### Fitting the CNN Model

```
import time

start = time.time()
print("hello")

# fitting the CNN Model
history = model.fit(
      train_generator,
      steps_per_epoch=10, #200 train images / 20 images per batch
      epochs=10,
      validation_data=validation_generator,
      validation_steps=5) #100 validation images / 20 images per batch

end = time.time()
print(end - start)
```

### Evaluating the CNN Model

#### Learning Curves: Training vs Validation

```
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc)) # number of epochs

# plotting the training and validation accuracy by epoch
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'orange', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

# plotting the training and validation loss by epoch
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'orange', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```



*   At the 10th epoch, the CNN model demonstrated superior performance on the validation dataset, achieving an impressive accuracy of 85% with a notably low loss value of 0.45.

* The validation data of the model outperformed the training data, which exhibited 79% accuracy and a slightly higher loss value of 0.49 at the same epoch, indicating that the model did not overfit to the training data.

* These results suggest that further training iterations may lead to improved accuracy on both training and validation datasets, potentially enhancing the model's overall performance and generalization capabilities of drones vs birds.

##Model 2 (with Dropout & Data Augmentation)

### Image Augmentation

```
datagen = ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
```

```
# importing modules for image preprocessing utilities
from keras.preprocessing import image
from google.colab import files

# picking a bird image from train partition image to "augment"
img_path = train_dir + '/birds/0255f71c-ae8d-4668-a9b8-90be8c96b2c2.jpg'

# Reading the image and resizing it
img = image.load_img(img_path, target_size=(180, 180))

# Converting it to a Numpy array with shape (180, 180, 3)
x = image.img_to_array(img)

# Reshapeing it to (1, 180, 180, 3)
x = x.reshape((1,) + x.shape)

# The .flow() command below generates batches of randomly transformed images.
# It will loop indefinitely, so we need to `break` the loop at some point!
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break

plt.show()
```

### Building the CNN Model

```
#building the cnn model
model = models.Sequential() #sequential api for model initiation
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(180, 180, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.1)) #adding a dropout layer before dense layers
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid')) #sigmoid activation for classification in output layer
```

### Compiling the CNN Model

```
from keras import optimizers

#compiling the model
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])
```

### Fitting the CNN Model

```
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

# Note that the validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(180, 180),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(180, 180),
        batch_size=20,
        class_mode='binary')

#fitting the model
history = model.fit(
      train_generator,
      steps_per_epoch=10,
      epochs=10,
      validation_data=validation_generator,
      validation_steps=5)
```

### Evaluating the CNN Model

#### Learning Curves: Training vs Validation

```
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc)) #number of epochs

# plotting the training and validation accuracy by epoch
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'orange', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

# plotting the training and validation accuracy by epoch
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'orange', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```



* The validation data reached the maximum accuracy of 71% at the 5th epoch and after which the model's performance is stable without any changes. But the training data accuracy started declining after 8th epoch where it reached 67.5%.

* The validation loss is gradually decreasing through out the epochs and lowest at 10th epoch for both training and validation but the validation data has the lower loss of 0.52 when compared to training loss of 0.57.

* These results suggest that further training iterations may not lead to improved accuracy on both training and validation datasets, also the model did not over fit on training data but lower accuracy indicates the models inability to generalize well between the classes drones vs birds.
